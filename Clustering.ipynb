```python
#lets write a function to get all venues in one neghibourhood
def getNearByVenues(latitude,longitude,radius):
    url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, VERSION, latitude, longitude, radius, LIMIT)
    results = requests.get(url).json()
    venues = results['response']['groups'][0]['items']
    nearby_venues = json_normalize(venues)
    filteredColumns = ['venue.name','venue.categories','venue.location.lat','venue.location.lng']
    nearby_venues = nearby_venues[filteredColumns]
    nearby_venues['venue.categories']=nearby_venues['venue.categories'].apply(lambda x: x[0]['name'])
    nearby_venues.rename(columns={'venue.categories': 'Category','venue.name':'Name','venue.location.lat':'Latitude','venue.location.lng':'Longitude'},inplace=True)
    return nearby_venues
```

# I ignore any neighbourhood which has less than 2 venues

```python
def getVenuesList(neighbourhoodList,latitudeList,longitudeList):
    VenuesList = pd.DataFrame(columns=['Neighbourhood','Longitude','Latitude','Venue Name','Venue Latitude','Venue Longitude','Venue Category'])
    INDEX = -1
    for neighbourhood,lat,lng in zip(neighbourhoodList,latitudeList,longitudeList):
        INDEX +=1
        try:
            
            nearbyVenues = getNearByVenues(latitude=lat,longitude=lng,radius=500)
            if nearbyVenues.shape[0]>1 :
                print(INDEX,'yes')
                for index,venue in nearbyVenues.iterrows():
                    VenuesList = VenuesList.append({
                    'Neighbourhood':neighbourhood,
                    'Longitude':lng,
                    'Latitude':lat,
                    'Venue Name':venue['Name'],
                    'Venue Latitude':venue['Latitude'],
                    'Venue Longitude':venue['Longitude'],
                    'Venue Category':venue['Category']}
                    ,ignore_index=True)
        except:
            print(INDEX)

    return VenuesList
    
        
    
```

# now lets do one hot encoding to analyze each neighbourhood


```python
torronto_onehot = pd.get_dummies(venuesList['Venue Category'])
# add the neighbourhood column
torronto_onehot['Neighbourhood'] = venuesList['Neighbourhood']
torronto_onehot
```

```python
columns = [torronto_onehot.columns[-1]]+list(torronto_onehot.columns[:-1])
torronto_onehot = torronto_onehot[columns]
torronto_onehot
```

```python
torronto_grouped = torronto_onehot.groupby('Neighbourhood').mean().reset_index()
torronto_grouped
```

```python
# lets make it a function which returns the first n most frequent venues
def retrieveNeighbourhoodsMostFrequentVenues(dataset,nom_top_venues):
    indicators = ['st','nd','rd']
    columns = []
    for i in np.arange(nom_top_venues):
        try:
            columns.append('{}{} most common venue'.format(i+1,indicators[i]))
        except:
            columns.append('{}th most common venue'.format(i+1))
    NeighbourhoodsMostFrequentVenues = pd.DataFrame(columns=columns)
    for hood in dataset['Neighbourhood'] :
        temp = dataset[dataset['Neighbourhood']==hood].T.reset_index()
        temp.columns = ['venue','freq']
        temp = temp.iloc[1:]
        temp['freq'] = temp['freq'].astype('float')
        temp = temp.sort_values('freq',ascending=False)[:nom_top_venues].T
        temp.columns = columns
        temp = temp.iloc[0]
        NeighbourhoodsMostFrequentVenues=NeighbourhoodsMostFrequentVenues.append(temp,ignore_index=True)
    NeighbourhoodsMostFrequentVenues['Neighbourhood'] = dataset['Neighbourhood']
    columns = [NeighbourhoodsMostFrequentVenues.columns[-1]]+list(NeighbourhoodsMostFrequentVenues.columns[:-1])
    NeighbourhoodsMostFrequentVenues = NeighbourhoodsMostFrequentVenues[columns]
    return NeighbourhoodsMostFrequentVenues
```

```python
TNMostFrequentVenues = retrieveNeighbourhoodsMostFrequentVenues(torronto_grouped,5)
TNMostFrequentVenues
```

# 5. Clustering using KMeans

```python
# now lets actually cluster neighbourhoods
def clusterNeighbourhood(dataset,nClusters):
    dataset_clustering = dataset.drop('Neighbourhood',1)
    kmeans = KMeans(n_clusters=nClusters,random_state=0,n_init=20).fit(dataset_clustering)
    return kmeans.labels_
```

```python
nClusters = 5
clusterLabels = clusterNeighbourhood(torronto_grouped,nClusters)
TNMostFrequentVenues['Cluster Labels'] = clusterLabels
```

```python
torronto_grouped['Cluster Labels'] = clusterLabels
#lets also add the longitude and latitude
torronto_grouped_withCoordinates = torronto_grouped.set_index('Neighbourhood').join(df[['Longitude','Latitude','Neighbourhood']].set_index('Neighbourhood'))
#lets move the last two columns here
columns = list(torronto_grouped_withCoordinates.columns[-2:])+list(torronto_grouped_withCoordinates.columns[:-2])
torronto_grouped_withCoordinates = torronto_grouped_withCoordinates[columns]
torronto_grouped_withCoordinates.reset_index(inplace=True)
torronto_grouped_withCoordinates
```

```python
torronto_grouped_cluster = torronto_grouped.groupby('Cluster Labels').mean()
torronto_grouped_cluster
```

# 6. Understanding whats is most frequent venues in each cluster

```python
# lets make it a function which returns the first n most frequent venues
def retrieveClustersMostFrequentVenues(dataset,nom_top_venues):
    indicators = ['st','nd','rd']
    columns = []
    for i in np.arange(nom_top_venues):
        try:
            columns.append('{}{} most common venue'.format(i+1,indicators[i]))
        except:
            columns.append('{}th most common venue'.format(i+1))
    clustersMostFrequentVenues = pd.DataFrame(columns=columns)
    clustersMostFrequentVenuesFreq = pd.DataFrame(columns=columns)

    for index,row in dataset.iterrows() :
        temp = pd.DataFrame([row.sort_values(ascending=False)[:nom_top_venues].index],columns=columns)
        temp1 = pd.DataFrame([row.sort_values(ascending=False)[:nom_top_venues]])
        temp1.columns = columns
        #temp.columns = ['venue','freq']
        #temp = temp.iloc[1:]
        #temp['freq'] = temp['freq'].astype('float')
        #temp = temp.sort_values('freq',ascending=False)[:nom_top_venues].T
       # temp.columns = columns
      #  temp = temp.iloc[0]
        clustersMostFrequentVenues=clustersMostFrequentVenues.append(temp,ignore_index=True)
        clustersMostFrequentVenuesFreq=clustersMostFrequentVenuesFreq.append(temp1,ignore_index=True)
        
        
    clustersMostFrequentVenues.index.name = 'Cluster Labels'
    clustersMostFrequentVenuesFreq.index.name = 'Cluster Labels'
    return clustersMostFrequentVenues,clustersMostFrequentVenuesFreq
```

```python
# now lets see what are the most common venues in each cluster, so that we can name the clusters
clustersFV,clustersFVFreq = retrieveClustersMostFrequentVenues(torronto_grouped_cluster,5)
clustersFV
```
