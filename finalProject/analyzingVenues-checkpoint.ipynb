```python
# lets now retrieve all of the table cells from the page
url = "https://en.wikipedia.org/wiki/List_of_cities_in_China_by_population"
page = requests.get(url)
# store the content of the website under doc
doc = lh.fromstring(page.content)
#parse data that are stored between <tr>..<tr>
tr_elements = doc.xpath('//tr')
```

```python
# as you see we are only interested on those rows which have 4 columns
# therefore I will creat a new list out of elements with the length of 9
elements4 = [T for T in tr_elements if len(T)==4]
i = 0 
# the first element with length of 4 is the columns heading
# note that we are only interested on province, capital and the population
col = ['Rank','Name','Province','Urban Population']
chinaCitiesAndPopulation = pd.DataFrame(columns = col)
chinaCitiesAndPopulation
```

```python
# now lets fill it
for t in elements4[1:]:
    chinaCitiesAndPopulation = chinaCitiesAndPopulation.append({
        'Rank':t[0].text_content().split('\n')[0]
        ,'Name':t[1].text_content().split('\n')[0]
        ,'Province':t[2].text_content().split('\n')[0]
        ,'Urban Population':t[3].text_content().split('\n')[0]  
    },ignore_index=True)
chinaCitiesAndPopulation
```

```python
# lets clean it up
import re
regex = re.compile(r"[,⍟#†*]")
chinaCitiesAndPopulation['Name'] = chinaCitiesAndPopulation['Name'].apply(lambda x: regex.sub('',x))
chinaCitiesAndPopulation['Urban Population'] = chinaCitiesAndPopulation['Urban Population'].apply(lambda x: regex.sub('',x))
chinaCitiesAndPopulation['Urban Population'] = chinaCitiesAndPopulation['Urban Population'].astype(int)
chinaCitiesAndPopulation
```

```python
# we define an index that show how many persons a city have per venue, which in turn shows that specific city demands more or less restaurants
personPerVenue = populationOfCutOffCities.values / venuesInEachCity.values
personPerVenue = pd.Series(personPerVenue,index=venuesInEachCity.index)
personPerVenue.sort_values(ascending = False)
```

```python
#lets wait here for a moment and assign an index to person per venue so that we have an indication of how good is a city to open a new business in (between 1 to 10)
#we define this new index to be
citySuccessIndex = ((personPerVenue.sort_values(ascending = False) - personPerVenue.sort_values(ascending = False)[-1])/(personPerVenue.sort_values(ascending = False)[0] - personPerVenue.sort_values(ascending = False)[-1]))*9+1
citySuccessIndex
```

```python
# since each city has a different number of venues lets take only 150 venues from each 
# Python code to generate 
# random numbers and 
# append them to a list 
import random 
# Function to generate 
# and append them  
# start = starting range, 
# end = ending range 
# num = number of  
# elements needs to be appended 
def Rand(start, end, num): 
    res = [] 
    j = -1
    
    while j<num-1:
        j+=1
        #create a random number
        x = random.randint(start, end)
        # check for duplicate
        if x in res:
            j-=1
        else:
            res.append(x) 
    return res
```

```python
venuesAfterRandomPicking = pd.DataFrame(columns =allVenuesAfterCutoff.columns )
for i in venuesInEachCity.iteritems():
    venues = allVenuesAfterCutoff[allVenuesAfterCutoff['City']==i[0]].reset_index(drop=True)
    res = Rand(start = venues.index.values[0],end = venues.index.values[-1],num = 150)
    venues = venues.loc[res,:]
    try:
        venuesAfterRandomPicking = venuesAfterRandomPicking.merge(venues,'outer')
    except:
        break
venuesAfterRandomPicking.reset_index(drop = True, inplace=True)
venuesAfterRandomPicking
```

```python
# now lets make a function that does this for each city 
def getProbaleSuccessfulVenues(city_name,data=venuesAfterRandomPicking):
    popularity_average = pd.get_dummies(data['Category'][data['City']!=city_name]).mean().sort_values(ascending=False)
    popularity_city = pd.get_dummies(data['Category'][data['City']==city_name]).mean().sort_values(ascending=False)
    return ((popularity_average - popularity_city)/popularity_average).fillna(0).sort_values(ascending=False)[:10]
```
