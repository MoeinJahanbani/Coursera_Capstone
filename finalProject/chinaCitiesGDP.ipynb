this notebook retrieves the population of each city in China

```python
import pandas as pd
import numpy as np
#lets import whatever is required
import requests
import lxml.html as lh
```

```python
# lets now retrieve all of the table cells from the page
url = "https://en.wikipedia.org/wiki/List_of_Chinese_administrative_divisions_by_GDP_per_capita?fbclid=IwAR0G_4iYC49v1e_USurmgF_PYjb1QpSOAf0qUMzopeHO3zogPKFLvZ89ekE"
page = requests.get(url)
# store the content of the website under doc
doc = lh.fromstring(page.content)
#parse data that are stored between <tr>..<tr>
tr_elements = doc.xpath('//tr')
```

```python
# as you see we are only interested on those rows which have 4 columns
# therefore I will creat a new list out of elements with the length of 9
elements4 = [T for T in tr_elements if len(T)==11]
i = 0 
# the first element with length of 4 is the columns heading
# note that we are only interested on province, capital and the population
col = ['Province','GDPpc(USD)']
GDPpc = pd.DataFrame(columns = col)
GDPpc
```

```python
# now lets fill it
numel = 0;
for t in elements4[0:]:
    numel+=1
    GDPpc = GDPpc.append({
        'Province':t[0].text_content().split('\n')[0]
        ,'GDPpc(USD)':t[2].text_content().split('\n')[0]  
    },ignore_index=True)
    if numel>31:
        break
GDPpc
```

```python
GDPpc.to_csv('GDPpc.csv',index=False)
```
